```python
import numpy as np

# Function to safely add derived features
def add_derived_features(df):
    df = df.copy()  # to avoid modifying original
    eps = 1e-5  # small constant to avoid division by zero

    # 1 HeightWidth interaction features
    df['Height_to_Width_Ratio'] = df['Equipment_Height'] / (df['Equipment_Width'] + eps)
    df['Height_x_Width'] = df['Equipment_Height'] * df['Equipment_Width']

    # 2 ValueWeight ratio
    df['Value_to_Weight_Ratio'] = df['Equipment_Value'] / (df['Equipment_Weight'] + eps)

    # (Optional) handle any inf or NaN created during division
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df.fillna(0, inplace=True)

    return df

# Apply to both train and test
X = add_derived_features(X)
test_df = add_derived_features(test_df)
X_train = add_derived_features(X_train)
X_test = add_derived_features(X_test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Base AdaBoost model
ada = AdaBoostRegressor(random_state=42)

# Parameter grid
param_grid_ada = {
    'n_estimators': [50, 100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'loss': ['linear', 'square', 'exponential']
}

# Grid Search
grid_search_ada = GridSearchCV(
    estimator=ada,
    param_grid=param_grid_ada,
    scoring='r2',
    cv=5,
    n_jobs=-1,
    verbose=2
)

# Train
grid_search_ada.fit(X_train, y_train.ravel())

# Best model
best_ada = grid_search_ada.best_estimator_

print("\n Best Parameters (AdaBoost):")
print(grid_search_ada.best_params_)

# Predict
y_pred_ada = best_ada.predict(X_test)

# Reverse transformation
y_pred_log = y_train_scaler.inverse_transform(y_pred_ada.reshape(-1, 1))
y_pred_final = np.expm1(y_pred_log).flatten()

# Evaluate
mse_ada = mean_squared_error(y_test_array, y_pred_final)
r2_ada = r2_score(y_test, y_pred_final)

print("\n AdaBoost Regressor (Optimized)")
print("MSE:", mse_ada)
print("R Score:", r2_ada)
```
